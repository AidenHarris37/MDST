```python
# write a function that takes a string as input
# and return that string in all caps

def return_upper(string: str) -> str:
    l = [c.upper() for c in string]
    s = ''.join(l)
    return s

print(return_upper('ahhhhhhhhhh'))
print(return_upper('hello mdst'))
```

```python
# fizzbuzz
# you will need to use both iteration and control flow 
# go through all numbers from 1 to 30 in order
# if the number is a multiple of 3, print fizz
# if the number is a multiple of 5, print buzz
# if the number is a multiple of 3 and 5, print fizzbuzz and NOTHING ELSE
# if the number is neither a multiple of 3 nor a multiple of 5, print the number

for i in range(1, 31):
    if i % 3 == 0 and i % 5 == 0:
        print('fizzbuzz')
    elif i % 3 == 0:
        print('fizz')
    elif i % 5 == 0:
        print('buzz')
    else:
        print(i)

    

```

```python
# load in the "starbucks.csv" dataset
# refer to how we read the cereal.csv dataset in the tutorial
import pandas as pd

df = pd.read_csv("/Users/aidenharris/Personal/MDST/2023-Tutorials/data/starbucks.csv")
df.head()
```

```python
# output the calories, sugars, and protein columns only of every 40th row. 

nt_facts = df.loc[::40, ['calories', 'sugars', 'protein']]
nt_facts.head()
```

```python
# select all rows with more than and including 400 calories
new_df = df[df['calories'] >= 400]
new_df.head()
```

```python
# select all rows whose vitamin c content is higher than the iron content
vit_c_df = df[df['vitamin c'] > df['iron']]
vit_c_df.head()
```

```python
# create a new column containing the caffeine per calories of each drink

df['caffeine per calories'] = df['caffeine'] / df['calories']
df.head()
```

```python
# what is the average calorie across all items?
import numpy as np
avg_calories = np.average(df['calories'])
print(avg_calories)
```

```python
# how many different categories of beverages are there?

unique_beverages = df['beverage_category'].unique()
num_unique_beverages = len(unique_beverages)
print(num_unique_beverages)
```

```python
# what is the average # calories for each beverage category?

avg_cals_by_category = {
    category: df[df['beverage_category'] == category]['calories'].mean()
    for category in unique_beverages
}
print(avg_cals_by_category)
```

```python
# plot the distribution of the number of calories in drinks with a histogram
import matplotlib.pyplot as plt

plt.hist(df['calories'])
plt.xlabel('calories')
plt.ylabel('frequency')
plt.show()
```

```python
# plot calories against total fat with a scatterplot
plt.scatter(df['total fat'], df['calories'])
plt.xlabel('total fat')
plt.ylabel('calories')
plt.show()
```
